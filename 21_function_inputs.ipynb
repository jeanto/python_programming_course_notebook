## 3. Mensagem Formal de *Obrigado*
- Modifique a função anterior para receber três argumentos — prefixo, primeiro nome e último nome — e imprimir uma mensagem de agradecimento (por exemplo, *Obrigado, Sra. Maria Lopez*):

```python
def print_obrigado_formal(prefixo, primeiro_nome, ultimo_nome): 
    """Imprime uma string contendo "Obrigado" e as entradas
    
    Parâmetros
    ----------
    prefixo: string
        Geralmente Sra., Sr.
    first_name: string
        Primeiro nome de uma pessoa
    last_name: string
        Último nome de uma pessoa
    """

    print("Obrigado", prefixo, primeiro_nome, ultimo_nome)
```

- Imprime mensagem formal de *Obrigado*: 

```python
for doador in potenciais_doadores:
    if doador[2] == "M":
        prefixo = "Sr."
    else:
        prefixo = "Sra."
    primeiro_nome = doador[0].split(' ')[0]
    ultimo_nome = doador[0].split(' ')[-1]
    # Se o primeiro e último nome forem iguais, não imprima o último nome
    #   (para evitar imprimir "Obrigado James James")
    #   porque o último nome não foi fornecido
    if (primeiro_nome == ultimo_nome):
        ultimo_nome = ""

    print_obrigado_formal(prefixo, primeiro_nome, ultimo_nome)
```

## 4. Último nome não fornecido! 
- Você percebe que alguns participantes não forneceram seus sobrenomes! Adapte a função para que o sobrenome tenha uma string vazia como valor padrão:

```python
def print_obrigado_default(prefixo, primeiro_nome, ultimo_nome=""): 
    """Imprime uma string contendo "Obrigado" e as entradas
    
    Parâmetros
    ----------
    prefixo: string
        Geralmente Sra., Sr.
    first_name: string
        Primeiro nome de uma pessoa
    last_name: string
        Último nome de uma pessoa. O padrão é uma string vazia
    """
    
    print("Obrigado", prefixo, primeiro_nome, ultimo_nome)
```

- Imprima mensagens de *Obrigado*, um com sobrenome e outro sem sobrenome:

```python
for doador in potenciais_doadores:
    if doador[2] == "M":
        prefixo = "Sr."
    else:
        prefixo = "Sra."
    primeiro_nome = doador[0].split(' ')[0]
    ultimo_nome = doador[0].split(' ')[-1]
    # Se o primeiro e último nome forem iguais, não imprima o último nome
    #   (para evitar imprimir "Obrigado James James")
    #   porque o último nome não foi fornecido
    if (primeiro_nome == ultimo_nome):
        print_obrigado_default(prefixo, primeiro_nome)
    else:
        print_obrigado_default(prefixo, primeiro_nome, ultimo_nome)
```
